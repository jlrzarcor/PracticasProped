```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from numpy import arange
```

```python
img=plt.imread('NArmsEAldMColl.png')
img.shape
img[:,:,0].shape
img = img[:,:,0] # Reducir al primer canal
imgplot = plt.imshow(img, cmap='gray')
plt.title('Imagen original modificada en escala de grises')
```

**Descomoposición de la Imágen por SVD**

```python
U,S,V_T = np.linalg.svd(img, full_matrices=False)
print('shapes U,:',U.shape,'S:',S.shape,'V:',V_T.shape)
rec_img_svd = (U*S)@V_T
rec_img_svd.shape
plt.imshow(rec_img_svd, cmap='gray')
plt.title('Imagen reconstruida utilizando SVD')
```

**1 VSI & 1 VSD & 1VS**

```python
rec_img_svd_1v=np.outer((U[:,1]*S[1]),V_T[1,:])
print('Shapes U*S;',(U[:,50]*S[50]).shape,'rec.ima. 1vec',rec_img_svd_1v.shape)
plt.imshow(rec_img_svd_1v, cmap='gray')
plt.title('SVD truncada a 1')
```

**2 VSI & 2 VSD & 2VS**

```python
rec_img_svd_2v=np.outer((U[:,1]*S[1]),V_T[1,:]) + np.outer((U[:,2]*S[2]),V_T[2,:])
plt.imshow(rec_img_svd_2v, cmap='gray')
plt.title('SVD truncada a 2')
```

**nVSI & nVSD & nVS**

```python
nvects=np.array([10,20,30,40,50,100,300])
rec_img_svd_10v=rec_img_svd_1v
rec_img_svd_20v=rec_img_svd_1v
rec_img_svd_30v=rec_img_svd_1v
rec_img_svd_40v=rec_img_svd_1v
rec_img_svd_50v=rec_img_svd_1v
rec_img_svd_100v=rec_img_svd_1v
rec_img_svd_300v=rec_img_svd_1v

for i in range(301):
    if i<=1:
        continue
    if i<=10:
        rec_img_svd_10v=rec_img_svd_10v+np.outer((U[:,i]*S[i]),V_T[i,:])
    if i<=20:
        rec_img_svd_20v=rec_img_svd_20v+np.outer((U[:,i]*S[i]),V_T[i,:])
    if i<=30:
        rec_img_svd_30v=rec_img_svd_30v+np.outer((U[:,i]*S[i]),V_T[i,:])
    if i<=40:
        rec_img_svd_40v=rec_img_svd_40v+np.outer((U[:,i]*S[i]),V_T[i,:])
    if i<=50:
        rec_img_svd_50v=rec_img_svd_50v+np.outer((U[:,i]*S[i]),V_T[i,:])
    if i<=100:
        rec_img_svd_100v=rec_img_svd_100v+np.outer((U[:,i]*S[i]),V_T[i,:])
    if i<=300:
        rec_img_svd_300v=rec_img_svd_300v+np.outer((U[:,i]*S[i]),V_T[i,:])
```
